**Finding max MSRP and the company name**

```
cd = max(car_data['MSRP'])
car_data['Make'][car_data['MSRP'] == cd]
```

**Finding the max MSRP each company has for its models**

```
msrp_data = car_data.groupby('Make')['MSRP'].max().reset_index()
msrp_data
```

**Dropping Irrelevant columns**

```
car_data = car_data.drop(['Market Category'], axis=1)
car_data.head()
```

**Renaming the Columns**

```
car_data = car_data.rename(columns = {"Engine HP": "HP", "Engine Cylinders": "Cylinders", "Transmission Type": "Transmission", "Driven_Wheels": "Drive Mode","highway MPG": "MPG-H", "city mpg": "MPG-C", "MSRP": "Price"})
car_data
```

**Plotting Price from the data set by using seaborn**

Finding max MSRP and the company name

```
cd = max(car_data['MSRP'])
car_data['Make'][car_data['MSRP'] == cd]
```

Finding the max MSRP each company has for its models

```
msrp_data = car_data.groupby('Make')['MSRP'].max().reset_index()
msrp_data
```

```
msrp_data1 = msrp_data.set_index('Make')
msrp_data1
```

```
msrp_data1.plot(kind = 'pie' , y = 'MSRP', figsize = [20,20])
```
